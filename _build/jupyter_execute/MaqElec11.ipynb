## Equacions dels transformadors 

$$ r_t = \frac{U_1}{U_2} = \frac{N_1}{N_2} $$

$$ P_2 = U_2 \cdot I_2 = \eta \cdot P_1 = \eta \cdot U_1 \cdot I_1 $$

### Transformadors ideals

$$ \eta = 1 \Rightarrow P_1 = P_2 \Rightarrow r_t = \frac{U_1}{U_2} = \frac{I_2}{I_1} $$
